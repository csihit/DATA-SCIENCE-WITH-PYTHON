# **Operators**

```python
#arithmetic Operators
a = 5
b = 3

# Addition
print("Addition (a + b):", a + b)

# Subtraction
print("Subtraction (a - b):", a - b)

# Multiplication
print("Multiplication (a * b):", a * b)

# Division
print("Division (a / b):", a / b)

# Remainder
print("Modulo (a % b):", a % b)

# a^b
print("Multiplication (a * b):", a ** b)
```

```python
# Relational Operators
a = 10
b = 5

# Equal to
print("Equal to (a == b):", a == b) # False

# Not equal to
print("Not equal to (a != b):", a != b) # True

# Greater than
print("Greater than (a > b):", a > b) # True

# Less than
print("Less than (a < b):", a < b) # False

# Greater than or equal to
print("Greater than or equal to (a >= b):", a >= b) # True

# Less than or equal to
print("Less than or equal to (a <= b):", a <= b) # False
```

```python
# Assignment Operators (Beginner Friendly)

# Start with a variable
x = 10
print("Initial value of x:", x)

# Simple Assignment
x = 5
print("After assigning 5, x =", x)

# Add and Assign
x += 3      # same as x = x + 3
print("After adding 3, x =", x)

# Subtract and Assign
x -= 2      # same as x = x - 2
print("After subtracting 2, x =", x)

# Multiply and Assign
x *= 4      # same as x = x * 4
print("After multiplying by 4, x =", x)

# Divide and Assign
x /= 2      # same as x = x / 2
print("After dividing by 2, x =", x)

# Remainder and Assign
x %= 3      # same as x = x % 3
print("After Remainder 3, x =", x)

# Exponent and Assign
x = 2
x **= 3     # same as x = x ** 3
print("2 power 3 =", x)
```

```python
# Logical Operators
a = True
b = False

# AND operator
print("a and b:", a and b) # False

# OR operator
print("a or b:", a or b) # True

# NOT operator
print("not a:", not a) # False
print("not b:", not b) # True
```

```python
# String for practice
text = "Data Science with Python"

# 1. length of string
print("Length:", len(text))

# 2. Convert to uppercase
print("Uppercase:", text.upper())

# 3. Convert to lowercase
print("Lowercase:", text.lower())

# 4. Capitalize first letter
print("Capitalized:", text.capitalize())

# 5. Replace word
print("Replace:", text.replace("Python", "AI"))

# 6. Find position of word
print("Find 'Science':", text.find("Science"))

# 7. Count word occurrence
print("Count of 'a':", text.count("a"))

# 8. Check if string starts with a word
print("Starts with 'Data':", text.startswith("pyhton"))

# 9. Check if string ends with a word
print("Ends with 'Python':", text.endswith("Python"))

# 10. Split string into list
print("Split words:", text.split())
```

```python
def print_list(lst):

    for item in lst:
        print(item, end=" ")

numbers = [1,2,3,4,5]

print_list(numbers)
```

```python
def factorial(n):

    fact = 1

    for i in range(1, n+1):
        fact *= i

    return fact

print(factorial(5))
```

```python
def list_length(lst):
    print(len(lst))

numbers = [10,20,30,40]

list_length(numbers)
```
