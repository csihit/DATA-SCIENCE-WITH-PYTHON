DATA VISUALIZATION

Bar Chart (Heart Disease Count)

```python
df['Heart Disease'].value_counts().plot(kind='bar')
plt.title("Heart Disease Count")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Number of People")
plt.show()
```

Pie Chart

```python
df['Heart Disease'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Heart Disease Percentage")
plt.ylabel("")  # remove y label
plt.show()
```

Histogram (Age Distribution)

```python
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Scatter Plot (Age vs Cholesterol)

```python
plt.scatter(df['Age'], df['Cholesterol'])
plt.title("Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()
```

Box Plot (Cholesterol)

```python
plt.boxplot(df['Cholesterol'])
plt.title("Cholesterol Distribution")
plt.show()
```

Simple Line Plot

```python
df['Age'].head(10).plot(kind='line')
plt.title("Age (First 10 People)")
plt.show()
```
