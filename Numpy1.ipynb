```python
import numpy as np

array1d = np.array([1, 2, 3, 4, 5])
print("1D Array:", array1d)

array2d = np.array([[10, 20, 30], [40, 50, 60]])
print("\n2D Array:\n", array2d)

zeros_array = np.zeros((3, 4))
print("\nZeros Array (3x4):\n", zeros_array)

ones_array = np.ones((2, 3))
print("\nOnes Array (2x3):\n", ones_array)

range_array = np.arange(0, 10, 2)
print("\nRange Array (0 to 10, step 2):", range_array)

linspace_array = np.linspace(0, 1, 5)
print("\nLinspace Array (5 points between 0 and 1):", linspace_array)
```

### Basic Array Arithmetic

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Addition:", a + b)

print("Subtraction:", a - b)

print("Element-wise Multiplication:", a * b)

print("Element-wise Division:", a / b)

print("Scalar Multiplication:", a * 3)

print("Dot Product:", np.dot(a, b))

print("Dot Product with @:", a @ b)
```

### Universal Functions (ufuncs)

```python
import numpy as np

array = np.array([-1, 0, 1, 2, 3.5])

print("Absolute values:", np.abs(array))

print("Square roots:", np.sqrt(array[array >= 0]))

print("Exponentials:", np.exp(array))

print("Natural Logarithms:", np.log(array[array > 0]))

print("Sine values:", np.sin(array))

print("Cosine values:", np.cos(array))

print("Sum of array elements:", np.sum(array))

print("Mean of array elements:", np.mean(array))

print("Maximum element:", np.max(array))

print("Minimum element:", np.min(array))
```
